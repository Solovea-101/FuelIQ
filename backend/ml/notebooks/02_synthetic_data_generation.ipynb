```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

df = pd.read_csv('../data/synthetic_trips.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Validate fuel consumption distributions
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0,0].hist(df['fuel_used_liters'], bins=50, edgecolor='black', alpha=0.7)
axes[0,0].set_xlabel('Fuel Used (Liters)')
axes[0,0].set_title('Distribution of Fuel Used')

axes[0,1].hist(df['distance_km'], bins=50, edgecolor='black', alpha=0.7, color='orange')
axes[0,1].set_xlabel('Distance (km)')
axes[0,1].set_title('Distribution of Trip Distance')

efficiency = df['distance_km'] / df['fuel_used_liters']
axes[1,0].hist(efficiency, bins=50, edgecolor='black', alpha=0.7, color='green')
axes[1,0].set_xlabel('Fuel Efficiency (km/L)')
axes[1,0].set_title('Distribution of Efficiency')
axes[1,0].axvline(efficiency.mean(), color='red', linestyle='--', label=f'Mean: {efficiency.mean():.1f}')
axes[1,0].legend()

axes[1,1].hist(df['avg_speed_kmh'], bins=50, edgecolor='black', alpha=0.7, color='purple')
axes[1,1].set_xlabel('Avg Speed (km/h)')
axes[1,1].set_title('Distribution of Average Speed')

plt.tight_layout()
plt.show()
```

```python
# Fuel consumption by route type and vehicle class
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

df['efficiency_kmpl'] = df['distance_km'] / df['fuel_used_liters']

sns.boxplot(data=df, x='route_type', y='efficiency_kmpl', ax=axes[0])
axes[0].set_title('Efficiency by Route Type')
axes[0].set_ylabel('Efficiency (km/L)')

sns.boxplot(data=df, x='vehicle_class', y='efficiency_kmpl', ax=axes[1])
axes[1].set_title('Efficiency by Vehicle Class')
axes[1].set_ylabel('Efficiency (km/L)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap for synthetic data
numeric_cols = ['engine_size_l', 'cylinders', 'distance_km', 'duration_minutes',
                'avg_speed_kmh', 'max_speed_kmh', 'idle_time_minutes',
                'load_weight_kg', 'fuel_used_liters']

plt.figure(figsize=(10, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='RdYlBu_r', center=0, fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

print('Correlations with fuel_used_liters:')
print(corr['fuel_used_liters'].sort_values(ascending=False))
```
